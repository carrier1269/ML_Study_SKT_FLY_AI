# Figure, Axis

```python
fig = plt.figure() # 여기서 현재는 figure는 객체만 가지고있음.
ax = fig.add_subplot() # 이렇게 부르면 Axes의 객체를 불러옴.
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 괄호안에는 axes의 위치를 넣음.
ax2 = fig.add_subplot(2,2,2) # 보통 python의 인덱스는 0부터 시작하는데,
ax3 = fig.add_subplot(2,2,3) # matplotlib의 인덱스는 1부터 시작한다.
ax4 = fig.add_subplot(2,2,4) # 첫 번째 axes를 수정하고싶다면, 변수지정해준 ax1를 컨트롤하면된다.


ax1.plot(x, y)
ax4.plot(x, y)
ax1.set_title('Graph')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221) # 괄호안에는 axes의 위치를 넣음.
ax2 = fig.add_subplot(222) # 보통 python의 인덱스는 0부터 시작하는데,
ax3 = fig.add_subplot(223) # matplotlib의 인덱스는 1부터 시작한다.
ax4 = fig.add_subplot(224) # 첫 번째 axes를 수정하고싶다면, 변수지정해준 ax1를 컨트롤하면된다.
# fig.add_subplot(221) = fig.add_subplot(2, 2, 1)

ax1.plot(x, y)
ax4.plot(x, y)
ax1.set_title('Graph')
plt.show()
```

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, color = '#00ff00', linewidth=5, linestyle = '-.') 
# 자주 쓰이는 색상은 'g'와 같이 가능
# 위와 같이도 표현가능, linewidth는 굵기, linestyle는 표시 형식이다.
plt.plot(x, z, color = 'orange') # 색상 풀네임을 넣어도 된다.
```

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, 'b--', marker='^')
# linestyle를 축약하여 'b--'와 같은 형식으로 할수도 있다.
# marker은 point부분에 표시형식을 수정한다.
plt.plot(x, z, label = 'Eng')
plt.legend(loc = 'lower right')
# label을 plot()에 추가하고 legend()명령을 주면 범례가 추가된다.
# legend()괄호에 loc, 범례 위치 설정가능.
```

### 막대그래프 barplot

```python
x = [i for i in range(5)]
y = [3, 2, 5, 4, 6]

plt.bar(x, y)
```

```python
x = ['A', 'B', 'C', 'D', 'E'] # 아래는 범주형 데이터
y = [3, 2, 5, 4, 6]

plt.bar(x, y, color = 'r') # plot과 같이 컬러변경가능.
```

### 히스토그램 (histogram)

```python
import numpy as np

mydata = np.random.randn(500) # 표준정규분포
# mydata
plt.hist(mydata) # 히스토그램은 x 축이 연속형데이터이다. 범주형데이터와 다른 것.
plt.show()
```

```python
import numpy as np

mydata = np.random.randn(500)
# mydata
plt.hist(mydata, bins = 20) # bins를 추가하면 구간을 나눌 수 있음.
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('iris.csv', index_col=0) # index_col=0, 0번째 컬럼을 인덱스로 설정한다.
df.shape
```

```python
plt.scatter(x = df['SepalLengthCm'],
            y = df['SepalWidthCm'], color = 'r')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('iris')
plt.show()
```

```python
plt.scatter(x = df.loc[df['Species'] == 'Iris-setosa','SepalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-setosa','SepalWidthCm'], color = 'r', label = 'Iris-setosa')
plt.scatter(x = df.loc[df['Species'] == 'Iris-versicolor','SepalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-versicolor','SepalWidthCm'], color = 'g', label = 'Iris-versicolor')
plt.scatter(x = df.loc[df['Species'] == 'Iris-virginica','SepalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-virginica','SepalWidthCm'], color = 'b', label = 'Iris-virginica')
plt.legend(loc = 'lower right')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('iris')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2) 

# ax1 plot 설정
ax1.scatter(x = df.loc[df['Species'] == 'Iris-setosa','SepalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-setosa','SepalWidthCm'], color = 'r', label = 'Iris-setosa')
ax1.scatter(x = df.loc[df['Species'] == 'Iris-versicolor','SepalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-versicolor','SepalWidthCm'], color = 'g', label = 'Iris-versicolor')
ax1.scatter(x = df.loc[df['Species'] == 'Iris-virginica','SepalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-virginica','SepalWidthCm'], color = 'b', label = 'Iris-virginica')
ax1.set_title('Sepal')

# ax2 plot 설정
ax2.scatter(x = df.loc[df['Species'] == 'Iris-setosa','PetalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-setosa','PetalWidthCm'], color = 'r', label = 'Iris-setosa')
ax2.scatter(x = df.loc[df['Species'] == 'Iris-versicolor','PetalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-versicolor','PetalWidthCm'], color = 'g', label = 'Iris-versicolor')
ax2.scatter(x = df.loc[df['Species'] == 'Iris-virginica','PetalLengthCm'],
            y = df.loc[df['Species'] == 'Iris-virginica','PetalWidthCm'], color = 'b', label = 'Iris-virginica')
ax2.set_title('Petal')
plt.show()
```

```python
mydata = [[1, 2, 3, 4],
          [10, 11, 12, 13],
          [30, 31, 32, 33],
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2,2)

z = 0

for x in range(2):
    for y in range(2):
        ax[x, y].plot(mydata[z])
        z += 1
```

```python
mydata = [[1, 2, 3, 4],
          [10, 11, 12, 13],
          [30, 31, 32, 33],
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2,2)

# 배열을 평탄화 시켜 축소작업을 할 수 있다. example> ax.flatten(), ax.replace(-1), ax.ravel()

ax = ax.reshape(-1)
for i in range(4):
    ax[i].plot(mydata[i])
```

```python
mydata = [[1, 2, 3, 4],
          [10, 11, 12, 13],
          [30, 31, 32, 33],
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2,2)

# 배열을 평탄화 시켜 축소작업을 할 수 있다. example> ax.flatten(), ax.replace(-1), ax.ravel()

ax = ax.reshape(-1)


for x, z in zip(ax, mydata): ############################################################
    x.plot(z)
```
