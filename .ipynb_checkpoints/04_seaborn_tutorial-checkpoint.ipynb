### Axes Level

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5,5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species')
# matplotlib의 scatter은 x, y만 대입하고 나머지는 직접 설정해줘야함.
# scatterplot의 hue는 분류 기준을 설정하는 컬럼값을 대입하면 된다.
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.histplot(data=df, x='species', hue='species')
```

```python
fig, axes = plt.subplots(1, 4, figsize = (12, 5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[0])
sns.histplot(data=df, x='species', hue='species', ax = axes[1])

sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[2])
sns.histplot(data=df, x='species', hue='species', ax = axes[3])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (12, 8))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[0, 0])
sns.histplot(data=df, x='species', hue='species', ax = axes[0, 1])

sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[1, 0])
sns.histplot(data=df, x='species', hue='species', ax = axes[1, 1])
plt.show()
```

### Figure-Level

```python
df.head()
```

```python
df.columns
```

```python
sns.displot(data=df, x='flipper_length_mm',
            kind="hist")
```

```python
sns.displot(data=df, x='flipper_length_mm',
            kind="hist", hue="species")
# hist, 히스토그램 / x축이 연속형 데이터인것.
```

```python
sns.displot(data=df, x='flipper_length_mm',
            kind="hist", hue="species", col="species")
plt.show()
```

```python
sns.displot(data=df, x='flipper_length_mm',
            kind="kde", hue="species")
```

```python
sns.displot(data=df, x='flipper_length_mm',
            kind="ecdf", hue="species")
```

```python
sns.relplot(data=df, x='flipper_length_mm',
            y='bill_length_mm',
            kind="scatter", hue="species", height=5, aspect=2) 
# height와 aspect를 이용하여 relplot, displot의 크기를 결정한다.
# 원래 matplot이나 seaborn에서 크기를 조정할때 기본으로
# plt.figure(figsize = (5,5))와 같은 식으로 크기를 조정하는데,
# relplot와 displot의 크기는 조정이 안된다. 왜나면 figure level이기 때문.
#aspect는 비율.
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
sns.barplot(data=df, x='Pclass', y='Fare')
```

```python
sns.countplot(data=df, x='Pclass')
plt.show()
```

```python
df['Pclass'].value_counts()
```

```python
sns.barplot(data=df, x='Pclass', y='Fare', hue='Sex', estimator=np.mean)
```

```python
sns.boxplot(data=df, x='Pclass', y='Age')
# 박스의 중앙선은 median값이다.
# 아래 선은 총 데이터 평균의 25%, 윗 선은 75%이다.
# T자를 이루고 있는 교차점은 윗면의 1.5배, 아랫면의 1.5배 수치이며,
# 이를 벗어나는 값의 데이터는 이상치(이상 데이터)라고 본다.
```

```python
sns.violinplot(data=df, x='Pclass', y='Age')
plt.show()
```

```python
data = np.random.rand(4, 3)
data
```

```python
sns.heatmap(data, annot = True, cbar=False, cmap='coolwarm') 
# 낮은 숫자일수록 검고, 큰 숫자일수록 하얗다.
# annot = True를 넣어주면 숫자 데이터를 같이 출력한다.
# cbar=False를 하면 우측 데이터바를 삭제함.
# cmap(colormap), color의 종류는 api에서 찾아보면 된다.
# 숫자가 낮을수록 coolwarm, cool톤, 높을수록 warm톤을 가지게 된다.
# 뒤에 오는 문자가 숫자 데이터가 높을수록 유사하게 가져가게 된다.
```

```python
df = pd.read_csv('tips.csv')
df
```

```python
df.groupby(['sex'])['tip'].value_counts()
```

```python
df['total_bill'].mean()
```

```python
sns.displot(data = df, x='tip', kind='hist', hue='sex', col='sex', height=5, aspect=0.5)
# 남여의 팁 분포도
```

```python
df.groupby(['sex'])['time'].value_counts() # 남여가 밥먹으러 오는 시간대
```

```python
df.groupby(['smoker'])['time'].value_counts() # 비/흡연자의 식사시간 조사
```

```python
df.groupby(['sex'])['smoker'].value_counts() # 남여의 흡연자수 조사
```

```python
df.loc[(df['sex']=='Female') & (df['smoker']=='No'),'tip'].value_counts()
# 여성이고 비흡연자인 사람의 팁 분포
```
