### Axes Level

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5,5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species')
# matplotlibì˜ scatterì€ x, yë§Œ ëŒ€ìž…í•˜ê³  ë‚˜ë¨¸ì§€ëŠ” ì§ì ‘ ì„¤ì •í•´ì¤˜ì•¼í•¨.
# scatterplotì˜ hueëŠ” ë¶„ë¥˜ ê¸°ì¤€ì„ ì„¤ì •í•˜ëŠ” ì»¬ëŸ¼ê°’ì„ ëŒ€ìž…í•˜ë©´ ëœë‹¤.
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.histplot(data=df, x='species', hue='species')
```

```python
fig, axes = plt.subplots(1, 4, figsize = (12, 5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[0])
sns.histplot(data=df, x='species', hue='species', ax = axes[1])

sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[2])
sns.histplot(data=df, x='species', hue='species', ax = axes[3])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (12, 8))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[0, 0])
sns.histplot(data=df, x='species', hue='species', ax = axes[0, 1])

sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax = axes[1, 0])
sns.histplot(data=df, x='species', hue='species', ax = axes[1, 1])
plt.show()
```

```python
df = pd.read_csv('titanic.csv')
df
```
